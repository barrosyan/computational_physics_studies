```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
wavelength=0.001
source_distance=1.0
screen_distance=2.0
```

```python
screen_position=np.linspace(-1.0,1.0,1000)
screen_position
```

```python
phase_difference=2*np.pi*source_distance*np.sin(screen_position) / wavelength
intensity=2 + 2* np.cos(phase_difference)
```

```python
phase_difference
```

```python
intensity
```

```python
plt.plot(screen_position,intensity)
plt.xlabel("Position on Observation Screen (m)")
plt.ylabel("Intensity")
plt.title("Interference Pattern of Two Coherent Sources")
plt.show()
```
