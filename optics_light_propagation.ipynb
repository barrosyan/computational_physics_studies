```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
focal_length=0.1
wavelength=0.5e-6
beam_radius=0.01
```

```python
input_plane=np.linspace(-0.02,0.02,1000)
input_plane
```

```python
lens_position=0
```

```python
output_plane=(1/(1-focal_length-1/(input_plane-lens_position)))+lens_position
output_plane
```

```python
output_intensity=(beam_radius**2)*(focal_length/(focal_length-(input_plane-lens_position)))**2
output_intensity
```

```python
plt.plot(input_plane,output_intensity)
plt.xlabel("Position at Input plane(m)")
plt.ylabel("Intensity at Output plane")
plt.title("Light propagation through lens")
plt.show()
```
