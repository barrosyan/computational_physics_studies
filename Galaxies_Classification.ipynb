```python
from random import randint

def gen_ips():
    network=ipaddress.IPv4Network("192.168.1.0/24")
    hosts=[]
    for host in network.hosts():
        hosts.append(host)
    return hosts
def scraping_data():
    dados=[]
    x=0
    y=0
    for data in nasa+eso:
        if 'galax' in data:
          host=hosts[y]
          network=ipaddress.IPv4Network(host)
          dados.append(requests.get(data).content)
          print(data)
          x+=1
          if x==8:
              y+=1
          sleep(randint(1,3))
    return dados
```

```python
imgsnasa=[]
for dado in dados:
  info=dado.decode('utf-8')
  if 'nasa' in info:
    while info != '':
      imgsnasa.append(info.partition('"href":"')[2].partition('"')[0])
      info=info.partition('"href":"')[2].partition('"')[2]
      print(dado)
```

```python
nasa_galaxies=[]
for img in imgsnasa:
  if '.json' in img:
    img=requests.get(img).content
    try:
      imgs=json.loads(img)
      for img in imgs:
        nasa_galaxies.append(img)
    except:
      nasa_galaxies.append(img)
  elif img != '':
    nasa_galaxies.append(img)
  else:
    pass
```

```python
nasa_galax=nasa_galaxies.copy()
nasa_galaxies=[]
for galaxie in nasa_galax:
  try:
    if '.jpg' in galaxie:
      nasa_galaxies.append(galaxie)
  except:
    galaxie=galaxie.decode('utf-8')
    if '.jpg' in galaxie:
      nasa_galaxies.append(galaxie)
```
