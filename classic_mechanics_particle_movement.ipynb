```python
#CLASSIC MECHANICS - TIME X SPACE - VELOCITY
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
```

```python
#DEFINING THE VARIABLES
t=np.linspace(0,10,100) #TIME
x=np.sin(t) #SPACE
v=np.cos(t) #VELOCITY - SPACE DERIVATE
```

```python
#CREATING THE TRAIN AND TEST DATA
X_train=t.reshape(-1,1)
Y_train=v.reshape(-1,1)

t_test=np.linspace(0,10,200)
x_test=t_test.reshape(-1,1)
```

```python
#CREATING AND FITING THE ML MODEL
model=MLPRegressor(hidden_layer_sizes=(10,),max_iter=1000)
model.fit(X_train,Y_train)
```

```python
#PREDICTING VELOCITY VECTOR WITH TEST DATA AND USING DERIVATE CONCEPTS TO CALCULATE X VECTOR
v_pred=model.predict(x_test)
x_pred=np.cumsum(v_pred)*(t_test[1]-t_test[0])
```

```python
#PLOTTING RESULTS
plt.figure()
plt.plot(t,x,label='Actual')
plt.plot(t_test,x_pred,label='Predicted')
plt.xlabel('Time')
plt.ylabel('Position')
plt.legend()
plt.show()
```
